# Import Packages

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Load Dataset

```python
confirmed = pd.read_csv('confirmed.csv')
recovered = pd.read_csv('recovered.csv')
death = pd.read_csv('death.csv')
```

```python
confirmed.head()
```

```python
recovered.head()
```

```python
death.head()
```

## Drop Columns (Data Cleaning)

```python
confirmed=confirmed.drop(['Province/State','Lat','Long'],axis=1)
```

```python
recovered=recovered.drop(['Province/State','Lat','Long'],axis=1)
```

```python
death=death.drop(['Province/State','Lat','Long'],axis=1)
```

```python
confirmed.head(10)
```

## GroupBy the Country Name

```python
confirmed = confirmed.groupby(confirmed['Country/Region']).aggregate('sum')
```

```python
confirmed.head(10)
```

```python
recovered = recovered.groupby(recovered['Country/Region']).aggregate('sum')
```

```python
recovered.head(10)
```

```python
death = death.groupby(death['Country/Region']).aggregate('sum')
```

```python
death.head()
```

## Transpose the Dataset

```python
confirmed=confirmed.T
death=death.T
recovered=recovered.T
```

```python
confirmed.head()
```

```python
confirmed.tail(10)
```

## Total New Cases

```python
new_cases = confirmed.copy()
for day in range(1,len(confirmed)):
    new_cases.iloc[day]=confirmed.iloc[day]-confirmed.iloc[day-1]
```

```python
confirmed.tail()
```

```python
new_cases.tail()
```

## Growth Rate

```python
growth_rate = confirmed.copy()
for day in range(1,len(confirmed)):
    growth_rate.iloc[day]=(new_cases.iloc[day]/new_cases.iloc[day-1])*100
```

```python
new_cases.tail()
```

```python
growth_rate.tail()
```

## Overall Growth Rate

```python
active = confirmed.copy()
for day in range(0,len(confirmed)):
    active.iloc[day]=confirmed.iloc[day]-death.iloc[day]-recovered.iloc[day]
overall_growth_rate = confirmed.copy()
for day in range(1,len(confirmed)):
    overall_growth_rate.iloc[day]=((active.iloc[day]-active.iloc[day-1])/active.iloc[day-1])*100
overall_growth_rate.tail()
```

## Death Rate

```python
death_rate=confirmed.copy()
for day in range(1,len(confirmed)):
    death_rate.iloc[day]=(death.iloc[day]/confirmed.iloc[day])*100
death_rate.tail()
```

## Data Vistualization

## Confirmed Cases

### Total Confirmed Cases in Countries

```python
countries=['India','Pakistan','Afghanistan','Italy']
ax=plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',color='white')
ax.tick_params(axis='y',color='white')
ax.set_title('COVID-19 Total Confirmed Cases by country',color='white')
for country in countries:
    confirmed[country][30:].plot(label=country)
plt.legend(loc='upper left')
plt.show()
```

### Bar Chart Plotting

```python
countries=['India','Pakistan','Afghanistan','Italy']
for country in countries:
    ax=plt.subplot()
    ax.set_facecolor('black')
    ax.figure.set_facecolor('#121212')
    ax.tick_params(axis='x',color='white')
    ax.tick_params(axis='y',color='white')
    ax.set_title(f'COVID-19 Total Confirmed Cases Growth Rate {country}',color='white')
    growth_rate[country].plot.bar()
    plt.show()
```

## Death Rate

```python
countries=['China','India','Pakistan','Afghanistan','Italy']
for country in countries:
    ax=plt.subplot()
    ax.set_facecolor('black')
    ax.figure.set_facecolor('#121212')
    ax.tick_params(axis='x',color='white')
    ax.tick_params(axis='y',color='white')
    ax.set_title(f'COVID-19 Total Confirmed Cases Death Rate {country}',color='white')
    death_rate[country].plot.bar()
    plt.show()
```
